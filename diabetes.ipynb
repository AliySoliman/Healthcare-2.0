   <b>Handling outliers using IQR</b>
</div>


```python
def check_outliers(col,br):
    q1 = br[col].quantile(0.25)
    q3 = br[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR
    outlier_in_df = br[(br[col] < lower_bound) | (br[col] > upper_bound)].shape[0]
    if outlier_in_df > 0:
        print(f'{col} contains outliers with count : {outlier_in_df}')
    else:
        print(f"{col} Does'nt contain outliers")    
```

```python
def remove_outliers(col, br):
    q1 = br[col].quantile(0.25)
    q3 = br[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR

    outlier_mask = (br[col] < lower_bound) | (br[col] > upper_bound)
    outlier_count = outlier_mask.sum()
    if outlier_count !=0:        
        br = br[~outlier_mask]  # Keep only rows that are NOT outliers
    return br,outlier_count
```

```python
for col in diabetes.select_dtypes(['int','float']).drop(['Id','Outcome'],axis = 1):
    check_outliers(col,diabetes)
```

```python
count = 1
while count != 0:
    count = 0  
    for col in diabetes.select_dtypes(['int','float']).drop(['Id','Outcome'], axis=1):
        diabetes, removed = remove_outliers(col, diabetes)
        count += removed
        print(f"{removed} outliers removed from {col}")
```

```python
plt.figure(figsize=(16, 12))  
sns.set(font_scale=1.2)

corr = diabetes.drop(columns='Id', errors='ignore').corr()

ax = sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    annot_kws={"size": 10},
    linewidths=0.5,
    linecolor='white',
    cbar_kws={"shrink": 0.8}
)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

plt.title("Correlation Heatmap", fontsize=18, pad=20)
plt.tight_layout()
plt.show()
```
